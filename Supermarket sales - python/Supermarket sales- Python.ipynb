```python
def data_Specs(supermarket):
    print("Initial Specs on your data =================== \n")
    print("Data Shape: ", supermarket.shape)
    print("\n---------------------------------------------\n")
    s = (supermarket.dtypes == 'object')
    Cat_cols = list(s[s].index) 
    print("Category Columns (",len(Cat_cols),") : \n",Cat_cols)
    print("\n---------------------------------------------\n")   
    s = (supermarket.dtypes != 'object')
    Nums_cols = list(s[s].index) 
    
    print("Numeric Columns (",len(Nums_cols),") : \n",Nums_cols)
    print("\n---------------------------------------------\n")
    print("Null Ratio per Column: \n")
    print(supermarket.isnull().sum()/len(supermarket)*100)
    print("\n---------------------------------------------\n")
    print("Summary Statistic for numeric columns: \n")
    print(supermarket.describe())
    print("\n---------------------------------------------\n")
```

```python
data_Specs(supermarket)
```

```python
def satisfaction_level(Rating):
    if Rating >=8:
        return 'High Satisfaction'
    elif Rating > 6 :
        return 'Average Satisfaction'
    else:
        return 'Below Satisfaction'
```

```python
supermarket['satisfaction_level']=supermarket['Rating'].apply(satisfaction_level)
satisfaction_level=supermarket['satisfaction_level'].value_counts()

satisfaction_level
```

```python
satisfaction_level.plot.pie( labels= satisfaction_level)
plt.show()
```
