```python
superstore_.plot.hist(alpha=.10,figsize=(5,5),color=(['blue','Red']))
plt.grid()
plt.show()
```

```python
superstore_.plot.box(figsize=(5,10),color=('orange'))
plt.title('Sales')
plt.show()
```

```python
superstore_.plot.scatter(x='Sales', y='Profit', alpha=0.5, color='blue', title='Sales vs Profit')
plt.grid(True)
plt.show()
```

```python
superstore_.Category.value_counts().plot.bar(title='Category')
plt.xlabel('Category Name')
plt.ylabel('Num Of Each Category')
plt.show()
```

```python
import seaborn as sns

sns.pairplot(superstore_[['Sales', 'Profit', 'Quantity']], palette='Set1')
plt.show()
```

```python
correlation_matrix = superstore_[['Sales', 'Profit', 'Quantity']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.show()
```
